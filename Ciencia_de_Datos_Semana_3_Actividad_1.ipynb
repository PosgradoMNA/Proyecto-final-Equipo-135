# **Parte 1: Fundamentos de bases de datos**





1.  ** Fundamentos de bases de datos y bases para ciencia de datos.**

*Para mi, mi primer contacto fue con los documentos de Excel, como bien son citados los documentos tienen cierta base "estructurada" que los permite ser manejados de una manera mucho mas eficiente, pero en muchos casos no le quita lo complicado al buscar, filtrar, remplazar, realizar operaciones, etcetera...*

  Una base de datos es un repositorio donde se almacenan diferentes tipos de datos provenientes de diferentes fuentes con el propósito de tener un rápido acceso a estos de manera seleccionada para cualquier tipo de análisis que sea requerido. Una característica de las bases de datos es que no cuentan con herramientas propias para realizar análisis en su mismo entorno, sino que los datos se deben extraer para después poderse procesar externamente, lo cual permite que su enfoque sea únicamente el de almacenar y facilitar la extracción de la información de la manera más eficiente posible y se asegura como resultado la calidad y consistencia de la misma.

  En ciencia de datos, es una tarea común la de localizar y recopilar datos de diferentes fuentes y bases de datos y reorganizarla en un conjunto de datos que sean relevantes para el caso de estudio que se está tratando y que sirva para contestar preguntas que nos lleven a los hallazgos con los que podemos fundamentar la toma de decisiones en una problemática específica del negocio. En otras ocasiones los datos puede ser que vengan de una sola base de datos, pero en un caso real no es lo más común, por la complejidad de variables que intervienen y que se desean medir.

  Una de las diferencias entre un analista de datos y un científico de datos es la habilidad de poder predecir datos en el futuro por medio de modelos que mejoran con el tiempo, mientras que un analista sólo observa datos históricos y la toma de decisiones basada en estos datos no le permite a la organización anticiparse a cambios en los datos reales y tener una ventaja competitiva con respecto a sus competidores.

# **Parte 2: Selección y limpieza de los Datos en Python**

```python
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv')
print(data)
```

```python
data.shape #nos explica como es la forma de nuestro Dataset
```

```python
data.describe()   #Para tener un resumen estadístico de los datos
```

```python
data.isnull().values.any() #nos dice si existen valores vacios
```

```python
data.isna().values.any() #nos dice si existen valores NA
```

```python
data.isnull().any() #nos dice si existen valores vacios en cada columna
```

```python
#iteramos para rellenar campos vacíos en columnas 6 a 11 con la moda o el valor más frecuente por columna
#estos valores pueden van de -1 a 9 y reflejan morosidad de los clientes

for i in range(6,12):   
  data.iloc[:,i].fillna(data.iloc[:,i].mode, inplace=True)
```

```python
data.isnull().any()   #Revisamos que ya no haya campos vacíos para variables X6 a X11
```

```python
data.dropna(inplace = True) # Asigna en los lugares vacios un "True" para remover los campos
```

```python
data # despliega el Dataset
```

Como podemos ver, ha cambiado el valor de los 30,000 a 29,971
