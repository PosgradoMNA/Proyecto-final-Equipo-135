## 1. Normalización - Min - Max

$$ x_i^{new} = \cfrac{x_i{old}-min(varX)/{range(varX)}$$


```python
def normalizarMinMax(x):
  min = np.min(x)
  max= np.max(x)
  range = max - min
  return [(xi - min)/
          range
          for xi in x
          ]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled = scaler.fit_transform(df)
scaled
```

```python
# creamos un data frame con los valores escalados para visualizar mejor la
# información:

scaled_df = pd.DataFrame(scaled, columns=df.columns)
scaled_df.head()
```

```python
# Repetimos el cálculo de las varianzas y valores min/max para el nuevo conjunto de datos

summary = {
"Varianzas": scaled_df[var].var(),
"Valor Min": scaled_df[var].min(),
"Valor Max": scaled_df[var].max()}

pd.DataFrame(summary, index = var).transpose()
```

Ahora los valores están en la misma escala y podemos proceder a aplicar PCA:

```python
from sklearn.decomposition import PCA
pcs = PCA()

pcs_t = pcs.fit_transform(scaled_df)
pcs_t
```

```python
# Calculamos el % de la varianza explicada por cada componente del PCA, así como
# El acumulado de la varianza explicada por cada componente

pcsSummary_df = pd.DataFrame({'Desviación estándar': 
                              np.sqrt(pcs.explained_variance_).round(4),
                              'Varianza explicada': 
                              pcs.explained_variance_ratio_.round(4),
                              'Varianza acumulada': 
                              np.cumsum(pcs.explained_variance_ratio_).round(4)})

pcsSummary_df
```

```python
# Creamos una lista con los nombres de cada componente: PC1, PC2 ... PC13

pcs_labels = [f'PC{i + 1}' for i in range(len(scaled_df.columns))]
pcsSummary_df.index = pcs_labels
pcsSummary_df
```

```python
# Creamos un dataframe con los valores o peso de cada componente.

pcs_df = pd.DataFrame(pcs_t, columns =pcs_labels)
print("Varianza total variables originales: ", scaled_df.var().sum())
print("Varianza total de los componentes: ", pcs_df.var().sum())
```

```python
pcs_df.head()
```

```python
pcsSummary_df
```

En este resumen podemos ver que la suma acumulada de las varianzas de los componentes PC1 al PC6 dan 81.15% de la varianza total de los componentes. Por lo tanto, si dejamos fuera los componentes del PC7 al PC14, aún mantenemos la mayor parte de la varianza total.

```python
cmpts_df = pd.DataFrame(
pcs.components_.round(4), # pcs.components nos da los pesos de cada uno de los componentes
columns = pcs_df.columns, # Cada columna muestra cada uno de los componentes.
index = scaled_df.columns) #Las filas = variables originales

cmpts_df.iloc[:,:6]
```

Ahora veamos cómo es la composición de cada componente arrojado por PCA. Solo nos quedaremos con los primeros 7 componentes, ya que estos explican la mayoría de la información.

De la información anterior podemos ver cómo cada variable original contribuye a los pesos de cada componente (en este caso nos interesan los valores absolutos). Obtengamos los valores máximos de cada componente para ver mejor la información.

```python
cmpts_df.iloc[:,:6].idxmax() #Nos interesa buscar los máximos por filas, 
                            #no por columnas. Por eso usamos idxmax
```
